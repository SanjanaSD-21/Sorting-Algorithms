BUBBLE SORT

```python
input_list = input("Enter a elements of the list: ")   # Accepting input from user
lst = input_list.split()                       
n = len(lst)
for i in range(0 , n):                              # Display the accepted input by coverting input to integer
    lst[i] = int(lst[i])
print("The input list is: ",lst) 
for i in range(0 , n - 1):                          # i is used to keep track of passess
    print('Pass:', i+1)
    for j in range(0 , n - i - 1):                  # j keeps track of comparison
        if(lst[j] < lst[j+1]):
            print('Comparison:', j + 1, 'NO SWAP')
        if(lst[j] > lst[j+1]):                     # swap if the number is greater tha its succeeding number
            print('Comparison:', j + 1, 'SWAP')
            temp = lst[j]
            lst[j] = lst[j+1]
            lst[j+1] = temp
            print(lst)
        
print("The sorted list is: " , lst)
```

INSERTION SORT

```python
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
for i in range(1 , n):                  # i keeps track of passes
    print('Pass:', i)
    temp = list[i]
    print("Value of temp is: ",temp)
    j = i - 1
    while j >= 0 and temp < list[j]: #comparison with temp of each element
        list[j+1] = list[j]           # shifting of position ahead
        j = j - 1
        print(list)
    list[j + 1] = temp
    print(list)
        
print("The sorted list is: " , list)
```

QUICK SORT

```python
def partition(list,low,high):  #partition logic
    pivot = low
    i = low
    j = high
    while(i<j):
        while(list[i]<=list[pivot] and i < high):
            i = i + 1
            #print('i=',i)
        while(list[j] >= list[pivot] and j > low):
            j = j - 1
            #print('j=',j)
        print("i = ", i)
        print("j = ", j)
        if(i < j):
            list[i],list[j] = list[j],list[i]
            print("After exchange i and jif i < j",list)
            
    list[j], list[pivot] = list[pivot], list[j]
    print("After complete pass: ",list)
    print("\n")
    return j
        
  # Function to do Quick sort 
def quickSort(list,low,high): 
    if low < high: 
        pi = partition(list,low,high) 
        # Separately sort elements before 
        # partition and after partition recursively.
        quickSort(list, low, pi-1) 
        quickSort(list, pi+1, high) 
  
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
quickSort(list,0, n-1)
print("The sorted list is: ",list)
```

MERGE SORT

```python
def Merge(list,low,mid,high):  #merging logic
    left_list = list[low:mid+1]
    print("List_1: ",left_list)
    n1 = len(left_list)
    right_list = list[mid+1:high+1]
    print("List_2: ",right_list)
    n2 = len(right_list)
    k = low
    i = 0
    j = 0
    while i < n1 and j < n2:
        if(left_list[i] <= right_list[j]):
            list[k] = left_list[i]
            i = i+1
            k = k+1
        else:
            list[k] = right_list[j]
            j = j+1
            k = k+1
            
    while(i < n1):
            list[k] = left_list[i]
            i = i+1
            k = k+1
    while(j < n2):
            list[k] = right_list[j]
            j = j+1
            k = k+1
    print("After merging above 2 list: ", list,"\n")
# Function to do Merge sort 
def MergeSort(list,low,high):
    if low < high: 
        mid = int((low + high)/2)
        print("Call mergeSort left")
        MergeSort(list, low, mid) #dividing the list into halves
        print("Call mergeSort right")
        MergeSort(list, mid+1, high) 
        print("\nNow Merge!!!")
        Merge(list, low, mid,high)
        
        
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
MergeSort(list,0, n-1)
print("The sorted list is:",list) 

```

HEAP SORT

```python
# heapify
def Heapify(list, n, i):
    largest = i # largest value
    l = 2 * i + 1 # left
    r = 2 * i + 2 # right
    if l < n and list[i] < list[l]:
        largest = l
    if r < n and list[largest] < list[r]:
        largest = r
    if largest != i:
        list[i],list[largest] = list[largest],list[i]
        Heapify(list, n, largest)
        
def HeapSort(list):
    n = len(list)
   # maxheap
    print("Building max heap.........")
    for i in range(n-1, -1, -1):
        Heapify(list, n, i)
        print(list)
   # deletion and extarction
    print("Heap ready!!!")
    print("Deleting and sorting elements.......")
    for i in range(n-1, 0, -1):
        list[i], list[0] = list[0], list[i] # swap
        Heapify(list, i, 0)
        print("After swap and heapify list is: ",list)
    print("Sorted!!!!")
# main
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
HeapSort(list)
print("The sorted list is:",list) 
```

RADIX SORT

```python
#counting Sort

def countingSort(list, exp): 
    n = len(list) 
    out = [0] * (n)  # Initialize lists to zero
    count = [0] * (10) 
  
    for i in range(0, n):        # find number of occurences, count list
        index = (list[i]//exp) 
        count[ (index)%10 ] += 1
    #print("Count array is: ",count)
  
    for i in range(1,10):      #cummulative count list
        count[i] += count[i-1] 
    #print("Cummulative Count array is: ",count)
  
    # Build the output array 
    i = len(list) - 1
    while i>=0: 
        index = (list[i]//exp) 
        out[ count[ (index)%10 ] - 1] = list[i] 
        count[ (index)%10 ] -= 1
        i -= 1
    print("Intermediate Output after evaluation of each place is:",out)
    i=0
    for i in range(0,len(out)):
        list[i]=out[i]
  
def RadixSort(list): 
    #Maximum number to know number of digits 
    max_num = max(list) 
    exp = 1
    while max_num//exp > 0: 
        countingSort(list,exp) 
        exp *= 10
  
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
RadixSort(list)
print("The sorted list is: ",list) 
```

SHELL SORT

```python
def ShellSort(list): 
    l = len(list) 
    gap = l//2        #create a large gap and the reduce
    while gap > 0: 
        if (gap == l//2):
            print("First Pass:")
        else:
            print("Next Pass:")
        for i in range(gap,l): 
            temp = list[i] 
            j = i 
            while  j >= gap and list[j-gap] >temp:  # comparing the elements at specific gaps
                list[j] = list[j-gap] 
                j = j - gap 
                print("After shifting",list)          #shifting the elements
            list[j] = temp 
            print("After shifting",list)
        gap = gap// 2                                 # reducing the gap
         
  
input_list = input("Enter a elements of the list: ")   # Accepting input from user
list = input_list.split()                       
n = len(list)
for i in range(0 , n):                   # Display the accepted input by coverting input to integer
    list[i] = int(list[i])
print("The input list is: ",list) 
ShellSort(list)
print("The sorted list is: ",list) 
  
```
